```python
import numpy as np
import pandas as pd
```

```python
latlong = pd.read_csv("/content/district wise centroids.csv")
census = pd.read_csv("/content/india-districts-census-2011.csv")
```

```python
cols = ["District code","State name","District name","Population","Male","Female","Literate",
        "Male_Literate","Female_Literate","Hindus","Muslims","Christians",
        "Sikhs","Buddhists","Jains","Others_Religions","LPG_or_PNG_Households","Housholds_with_Electric_Lighting",
"Households_with_Internet",
"Households_with_Computer"]
```

```python
censusf = census[cols]
```

```python
censusf["State name"] = censusf["State name"].replace({'NCT OF DELHI':"DELHI"})
```

```python
np.sort(censusf["State name"].unique())
```

```python
np.sort(latlong["State"].unique())
```

```python
state = {}
num=0
for i in np.sort(censusf["State name"].unique()):
  if i not in state:
    state[i] = np.sort(latlong["State"].unique())[num]
  num=num+1
print(state)
```

```python
censusf["State name"].replace(state,inplace=True)
```

```python
df = latlong.merge(censusf,left_on=["State","District"],right_on=["State name","District name"])
```
